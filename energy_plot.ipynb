```python
"""
Authors: Viktoriia Vlasenko
Index number: 317013
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def processFile(file_input):
    col_names = ["t", "H", "Ek", "V" ,"T", "P"]
    df = pd.read_csv(file_input, delim_whitespace=True, names = col_names)

    t = []
    for i in range(df.shape[0]):
        t.append(float(df['t'][i]))

    ec = []
    for i in range(df.shape[0]):
        ec.append(float(df['H'][i]))
        
    ek = []
    for i in range(df.shape[0]):
        ek.append(float(df['Ek'][i]))

    epot = []
    for i in range(df.shape[0]):
        epot.append(float(df['V'][i]))

    temp = []
    for i in range(df.shape[0]):
        temp.append(float(df['T'][i]))

    press = []
    for i in range(df.shape[0]):
        press.append(float(df['P'][i]))

    return [t, ec, ek, epot, temp, press]


if __name__ == '__main__':
    file_input1 = "total_energies.txt"

    time = np.array(processFile(file_input1)[0])
    e_c = np.array(processFile(file_input1)[1])

    e_k = np.array(processFile(file_input1)[2])
    e_pot = np.array(processFile(file_input1)[3])
    
    temp = np.array(processFile(file_input1)[4])
    press = np.array(processFile(file_input1)[5])

    fig, ax = plt.subplots( dpi=150, figsize=(8, 6))
    ax.plot(time, e_c,
            label="E Total")
    
    ax.plot(time, e_k,
             label="E Kinetic")
    
    ax.plot(time, e_pot,
            label="E Potential")
    
    ax.plot(time, temp,
            label="Temperature")
    
    ax.plot(time, press,
            label="Pressure")
    
    ax.set_xlabel("time [ps]")
    # ax.set_ylabel("Energy")
    ax.legend()
    ax.set_title("Energy , Temperature, Pressure plots for n = 6 and T = 1000K")
    #ax.set_title("Energy plots for n = 8 and T = 100K")
    plt.savefig(f"etotal_plots_for_n=6_T=1000.jpg", dpi=150)
    #plt.savefig(f"energies_plots_for_n=8_T=100.jpg", dpi=150)
    plt.show()
    
```
